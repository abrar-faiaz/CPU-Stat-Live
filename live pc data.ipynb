```python
import time
import psutil
import matplotlib.pyplot as plt
import numpy as np # import numpy
%matplotlib notebook
plt.rcParams['animation.html'] = 'jshtml'
plt.style.use('fivethirtyeight')

fig = plt.figure()
ax1 = fig.add_subplot(311) # create a subplot for cpu usage
ax2 = fig.add_subplot(312) # create a subplot for memory usage
ax3 = fig.add_subplot(313) # create a subplot for disk usage
fig.show()
i = 0
x, y1, y2, y3 = np.array([]), np.array([]), np.array([]), np.array([]) # use numpy arrays instead of lists

while True:
    x = np.append(x, i) # append to numpy array
    y1 = np.append(y1, psutil.cpu_percent()) # get the cpu percentage and append to numpy array
    y2 = np.append(y2, psutil.virtual_memory().percent) # get the memory percentage and append to numpy array
    y3 = np.append(y3, psutil.disk_usage('/').percent) # get the disk percentage and append to numpy array
    
    ax1.plot(x, y1, color='b') # plot the cpu usage
    ax2.plot(x, y2, color='g') # plot the memory usage
    ax3.plot(x, y3, color='r') # plot the disk usage
    
    fig.canvas.draw()
    
    ax1.set_xlim(left=max(0, i-50), right=i+50)
    ax2.set_xlim(left=max(0, i-50), right=i+50)
    ax3.set_xlim(left=max(0, i-50), right=i+50)
    
    ax1.set_xlabel('Time') # add x label for cpu usage
    ax1.set_ylabel('%') # add y label for cpu usage
    ax1.set_title('CPU Percentage') # add title for cpu usage
    
    ax2.set_xlabel('Time') # add x label for memory usage
    ax2.set_ylabel('%') # add y label for memory usage
    ax2.set_title('Memory Percentage') # add title for memory usage
    
    ax3.set_xlabel('Time') # add x label for disk usage
    ax3.set_ylabel('%') # add y label for disk usage
    ax3.set_title('Disk Percentage') # add title for disk usage
    
    fig.tight_layout() # adjust the spacing between subplots
    
    time.sleep(0.1)
    i += 1
plt.close()
```
